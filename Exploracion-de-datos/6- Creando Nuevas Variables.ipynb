## Creando Nuevas Variábles

La directoría de la empresa precisa nuevas informaciones, por lo que tendremos que reorganizar nuestro conjunto de datos para obtener las siguiente informaciones:

- Valor bruto del Alquiler
- Valor por metro cuadrado de un inmueble
- Agrupar los datos de Casas y Apartamentos

```python
import pandas as pd
```

```python
datos = pd.read_csv('datos/alquiler_residencial.csv', sep = ';')
```

```python
datos.head(10)
```

```python
datos['Valor Bruto'] = datos['Valor'] + datos['Condominio'] + datos['IPTU']
```

```python
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor'] / datos['Area']
```

```python
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor m2'].round(2)
```

```python
datos.head(10)
```

```python
datos['Valor Bruto m2'] = (datos['Valor Bruto'] / datos['Area']).round(2)
```

```python
datos.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
datos['Tipo Agregado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
datos
```

## Excluyendo Variábles

La variable IPTU tiene valores poco confiables, después de todo, el IPTU no siempre tiene valores declarados. Al final, la empresa decidió que el valor bruto y el valor bruto m2 deberían excluirse de la base de datos.

```python
# Creando un DataFrame para almacenar datos auxiliares
datos_aux = pd.DataFrame(datos[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
datos_aux.head(10)
```

```python
# Para eliminar una variable de la base de datos, usaremos el 'del', 
# pasaremos el DataFrame y el nombre de la variable que se eliminará.
del datos_aux['Valor Bruto']
```

```python
datos_aux.head(10)
```

```python
# Otra manera de excluir variábles es utilizando el método pop()
datos_aux.pop('Valor Bruto m2')
```

```python
datos_aux
```

```python
# Otra manera de excluir múltiples variábles es utilizando el método drop()
datos.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
datos.head(10)
```

```python
datos.to_csv('datos/alquiler_residencial.csv', sep = ';', index = False)
```
